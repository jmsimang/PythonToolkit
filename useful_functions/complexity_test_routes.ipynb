```python
import os
import sys
import pandas as pd
os.getcwd()
%matplotlib inline
sys.path.append('./recursion/')
sys.path.append('.')
```

```python
from useful_functions.counting_routes_combinations import count_routes_combinations as crc
from useful_functions.counting_routes import count_routes_iterative as cri
from useful_functions.recursion.counting_routes_recursive_lookup import counting_routes_r as crrl
from useful_functions.recursion.counting_routes_recursive import counting_routes_recursive as crr
```

```python
number = [2, 5, 10, 20, 50]
output = []
for n in number:
    out_combin = %timeit -o -q crc(n)
    out_iter = %timeit -o -q cri(n, n)
    out_recur_lu = %timeit -o -q crrl(n, n)
    if n <= 10:
        out_recur = %timeit -o -q crr(n, n)
    else:
        out_recur = None
    output.append([n, out_combin.best, out_iter.best, out_recur_lu.best, out_recur.best if n <= 10 else None])
```

```python
output = pd.DataFrame(output, columns=['Grid Size', 'Combinations', 'Iterative', 'Recursive Lookup', 'Recursive'])
output.set_index('Grid Size', inplace=True)
```

```python
output.plot(title='Time taken to solve Lattice Paths of various sizes')
```
